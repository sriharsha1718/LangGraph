```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
  message : str

def greeting_node(state: AgentState) -> AgentState:
  """Simple Node that adds a greeting message to the state"""
  state['message'] = 'Hey '+ state['message'] + ', How is your day going?'
  return state
```

```python
graph = StateGraph(AgentState)

graph.add_node('greeter', greeting_node)
graph.set_entry_point('greeter')
graph.set_finish_point('greeter')

app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'message':'Sri Harsha'})
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph

class AgentState(TypedDict):
  message : str

def compliment_node(state: AgentState) -> AgentState:
  """Adds the user name to a compliment message"""
  state['message'] = state['message'] + ', you are doing an amazing job learning LangGraph!'
  return state

graph = StateGraph(AgentState)

graph.add_node('complimenter', compliment_node)
graph.set_entry_point('complimenter')
graph.set_finish_point('complimenter')

app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'message':'Sri Harsha'})
```
